```python
import pandas as pd
import numpy as np
```

```python
# 시도 지명 바꾸기
def city_sido_change(sido):
    if sido =='서울':
        sido = '서울특별시'
    elif sido =='인천':
        sido = '인천광역시'
    elif sido =='경기':
        sido ='경기도'
        
    return sido
```

```python
# 오류 행정동 코드 바꾸기
def SGGcode_change(SGG_CD):
    if(SGG_CD=='23030'):
        SGG_CD='23090'
    else:
        SGG_CD
    
    return SGG_CD
```

```python
# EQ5D, 소득, 교육 PCA => 사회불평등 지수(독립변인)
socio_index = pd.read_excel('../data/temporary_storage/1_soc_index.xlsx')
socio_index = socio_index.astype({'행정구역코드':str})

socio_index['시도'] = socio_index['시도'].apply(lambda x:city_sido_change(x))
socio_index['행정구역코드'] = socio_index['행정구역코드'].apply(lambda x:SGGcode_change(x))

socio_index
```
